> La fuerza de 120[N] se aplica como se muestra en un extremo de la llave curva. Si α=30^o, calcule el momento de F con respecto al centro O del perno. <br> Determine el valor de α que maximizaría el momento con respecto a O, indique el valor de este momento máximo.

### DCL O

![title](E-sample2-58_dcl.png)

<br> Deducción de función del momento respecto a α

$$M_O^F=34,8 cos⁡(α)+22,8sen(α) $$

```python
th = np.arange(0,90,1)
vth = np.array(np.deg2rad(th))
R = (34.8*np.cos(vth)+22.8*np.sin(vth))
#plot
plt.scatter(th,R)
plt.title('Momento máximo')
plt.xlabel('alpha[deg]')
plt.ylabel('Momento[Nm]')
plt.show
```
