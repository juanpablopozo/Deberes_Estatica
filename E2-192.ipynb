> El brazo AB rota en el rango de $0 \leq \theta \leq 180^o$. El resorte se encuentra en su longitud natural como  $\theta=90^o$. Determine como una función de θ el momento alrededor de O producido por la fuerza elástica aplicada en B. Graficar las componentes X,Y,Z de M_O y encuentre el valor máximo absoluto de cada componente.

<br> Deducción de función del momento de la fuerza elástica respecto a α

$$M_O^F=216\sqrt{(8,64-2,88sen(\theta))} $$

```python
th = np.arange(0,180,1)
vth = np.array(np.deg2rad(th))
R = (216*np.sqrt(8.64-2.88*np.sin(vth)))
#plot
plt.plot(th,R)
plt.title('Momento máximo')
plt.xlabel('theta[deg]')
plt.ylabel('Momento[Nm]')
plt.show     
```
